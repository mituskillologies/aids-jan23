```python
sns.kdeplot(df['sepal_width'])
```

```python
sns.kdeplot(df['petal_length'])
```

```python
# Jointplot
sns.jointplot(x='sepal_length', y='petal_width', 
              data=df, hue='species')
```

```python
# Multivariate 
sns.pairplot(df, hue='species')
```

```python
# Count plot
sns.countplot(x=df['species'])
```
