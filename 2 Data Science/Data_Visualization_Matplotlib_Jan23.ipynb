### Import matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Declare the arrays

```python
x = np.array([3,8,13,22,27,38,45,49,59,81])
y = np.array([10,14,21,34,41,57,62,88,92,99])
z = np.array([87,81,77,76,54,41,21,18,17,16])
```

### Line Plot

```python
plt.plot(x, y)
```

```python
plt.plot(x, z)
```

```python
p = np.array([56,54,55,57,51,55,55,54,55,59])
```

```python
plt.plot(x, p)
```

```python
plt.grid()
plt.title('My first plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xlim(0,100)
plt.ylim(0,100)
plt.xticks(range(0,101,10))
plt.yticks(range(0,101,10))
plt.plot(x, y)
```

```python
plt.grid()
plt.xlabel('First Value')
plt.ylabel('Second Value')
# Marker: .  o  *   >    <    v   ^  D
# color: r g b / golden maroon purple / #78A567
# ms - marker size, mfc - marker face color, ls - line style
# lw - line width
plt.plot(x, y, marker = 'o', color = 'red', ms= 10, mfc='y',
        lw = 3, ls = ':')
```

```python
plt.figure(figsize=(16,9))
plt.grid()
plt.plot(x, y, color = 'r')
```

### Subplots

```python
plt.figure(figsize=(16,9))
plt.subplot(2,1,1)
plt.title('First')
plt.grid()
plt.plot(x,y, color='r', marker = 'o')

plt.subplot(2,1,2)
plt.title('Second')
plt.grid()
plt.plot(x,z, color='g', marker = '*')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(x,p)
plt.subplot(1,2,2)
plt.plot(x,z)
```

### Save the plot

```python
plt.figure(figsize=(16,9))
plt.subplot(2,1,1)
plt.title('First')
plt.grid()
plt.plot(x,y, color='r', marker = 'o')

plt.subplot(2,1,2)
plt.title('Second')
plt.grid()
plt.plot(x,z, color='g', marker = '*')
plt.savefig('myplot.png')
```

### Bar plot

```python
stud = [56,78,120,156,104,122]
lang = ['C','C++','Java','Python','PHP','JavaScript']
```

```python
plt.title('First Bar Plot')
plt.xlabel('Language Courses')
plt.ylabel('No. of Students')
plt.grid()
plt.bar(lang, stud, color = 'brown')
```

### Horizontal Bar Plot

```python
plt.title('First Bar Plot')
plt.xlabel('Language Courses')
plt.ylabel('No. of Students')
plt.grid()
plt.barh(lang, stud, color = 'brown')
```

### Pie chart

```python
plt.figure(figsize=(8,8))
plt.pie(stud, labels=lang, autopct='%2.2f%%');
```

# Univariate Analysis

### Histogram

```python
d = [45,44,34,67,87,54,51,25,43,23,56,65,45,61,62,55,34,58]
```

```python
plt.xticks(range(0,101,10))
plt.grid()
plt.title('My Histogram')
plt.xlabel('Ranges')
plt.ylabel('Count')
plt.hist(d, bins=range(0,101,10), color = 'darkgreen');
```

```python
plt.boxplot(marks);
```

### Comparing the plot

```python
plt.plot(x, y, color = 'red', label = '2022')
plt.plot(x, z, color = 'green', label = '2023')
plt.grid()
plt.legend()
```
